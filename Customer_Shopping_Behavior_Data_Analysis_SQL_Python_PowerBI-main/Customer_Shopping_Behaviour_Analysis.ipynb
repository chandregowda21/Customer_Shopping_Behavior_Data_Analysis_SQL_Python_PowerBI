```python
## Read the dataset
import pandas as pd

df = pd.read_csv('customer_shopping_behavior.csv')
df.head()
```

```python
## Display basic information about the dataset
df.info()
```

```python
## Display statistical summary of the dataset
df.describe(include='all')
```

```python
## Check for missing values
df.isnull().sum()
```

```python
## We will replace missing values in 'Review Rating' with the median rating of that product category
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
df.isnull().sum()
```

```python
## Rename columns for better readability and consistency
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
df.head()
```

```python
## Create a column age_group based on age
labels = ['Young Adult', 'Adult', 'Middle Aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
df[['age', 'age_group']].head(10)
```

```python
## Create column purchase_frequency_days 
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
df[['frequency_of_purchases', 'purchase_frequency_days']].head(10)
```

```python
## Display first 10 rows of discount_applied and promo_code_used columns
df[['discount_applied', 'promo_code_used']].head(10)
```

```python
## Compare two columns discount_applied and promo_code_used
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
## Drop promo_code_used column as both columns are identical
df = df.drop(columns=['promo_code_used'])
df.columns
```
